```python
import numpy as np
from keras.layers import Input, Dense
from keras.models import Model
import pandas as pd
```

```python
# Load and preprocess data
data =  pd.read_csv('dailytest.csv', index_col='Date',parse_dates=True)
data.index.freq = 'D'
```

```python
# Preprocess the data as needed (e.g., scaling, normalization, feature engineering)
# Convert date column to datetime object
#data['date'] = pd.to_datetime(data['date'])

# Set date column as index
#data.set_index('date', inplace=True)

# Fill missing values (if any) using linear interpolation
data.interpolate(method='linear', inplace=True)

```

```python
from sklearn.preprocessing import StandardScaler
# Scale the data using standardization
scaler = StandardScaler()
# scaled_data = scaler.fit_transform(data)
```

```python
data = data['Consumption'].values 
scaled_data = scaler.fit_transform(data.reshape(-1, 1))
```

```python
scaled_data
```

```python
# Split data into training and testing sets
train_data = scaled_data[:len(data)//2]
test_data = scaled_data[len(data)//2:]
```

```python
train_data
```

```python
test_data
```

```python
# Define the autoencoder model
input_layer = Input(shape=(train_data.shape[1],))
encoded = Dense(64, activation='relu')(input_layer)
encoded = Dense(32, activation='relu')(encoded)
decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(train_data.shape[1], activation='linear')(decoded)
```

```python
input_layer
```

```python
autoencoder = Model(input_layer, decoded)
autoencoder.compile(optimizer='adam', loss='mse')
```

```python
# Train the autoencoder model
autoencoder.fit(train_data, train_data,
                epochs=50,
                batch_size=32,
                )

# Use the autoencoder model to detect anomalies
predictions = autoencoder.predict(data)
mse = np.mean(np.power(data - predictions, 2), axis=1)
threshold = np.percentile(mse, 95)
anomalies = np.where(mse > threshold)[0]
```
